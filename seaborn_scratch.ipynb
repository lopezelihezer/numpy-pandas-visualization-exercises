```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')

tips.dtypes
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips) # x and y set as string names of columns
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
plt.title('Total Bill vs Tip Amount')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
```

```python
sns.relplot(x='total_bill', y='tip', col='day', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', hue='day', col='time', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', col='day', data=tips, kind='line')
```

```python
sns.histplot(tips.total_bill)
plt.show()
```

```python
sns.boxplot(data=tips, x='tip')
```

```python
sns.histplot(tips.tip)
```

```python
sns.catplot(data=tips, y='day', x='tip', hue='smoker', col='time')
```

```python
sns.dogplot()
```

```python
data = pd.crosstab(tips.time, tips['size'])
data
```

```python
sns.heatmap(data, annot=True, cmap=plt.cm.Greens)
```

```python
sns.pairplot(tips)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip')
```
