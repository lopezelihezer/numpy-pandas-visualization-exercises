```python
import matplotlib.pyplot as plt
print(plt)
```

```python
x = list(range(150))

plt.plot(x)
plt.show()
```

```python
plt.plot(range(len(x)), x)
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 5, 6])
```

```python
plt.xlim(-20, 200)
plt.ylim(-5, 160)
plt.plot()
```

```python
x = [1, 1, 1, 3, 2, 3, 3, 3, 4, 4, 6]
plt.hist(x)
plt.show()
```

```python
x = [1, 1, 1, 3, 2, 3, 3, 3, 4, 4, 6]
plt.hist(x, bins = [0, 2, 4, 6], edgecolor = 'black')
plt.show()
```
