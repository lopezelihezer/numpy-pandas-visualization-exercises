```python
import matplotlib.pyplot as plt
from random import randint
import numpy as np
import pandas as pd
import math
```

```python
plt.figure(figsize=(12, 5))

# make a list of x values
x = pd.Series(range(1, 10))
# create the different y plots
y1 = 0*x + 1 
y2 = np.log(x) #'O(log n)'
y3 = x #'O(n)'
y4 = x.apply(lambda x: x * np.log(x)) #'O(n log n)'
y5 = x**2 #'O(n^2)'
y6 = 2**x # 'O(2^n)'
y7 = x.apply(lambda x: math.factorial(x))# 'O(n!)'
y8 = x**x # 'O(n^n)'


plt.plot(x, y1, label= "O(1)")
plt.plot(x, y2, label='O(log n)')
plt.plot(x, y3, color='green', label='O(n)')
plt.plot(x, y4, color='yellow', label='O(n log n)')
plt.plot(x, y5, color='purple', label='O(n^2)')
plt.plot(x, y6, color='black', label='O(2^n)')
plt.plot(x, y7, color='cyan', label='O(n!)')
plt.plot(x, y8, color='pink', label='O(n^n)')

plt.ylim(0, 20)
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')

plt.legend()
plt.show()
```
