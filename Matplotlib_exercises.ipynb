```python
import numpy as np
import pandas as pd

x = pd.Series(range(0, 50)) # or some other range of numbers for your x values
def squared_minus_x_plus_two(x):
    return x**2 - x + 2
y = x.apply(squared_minus_x_plus_two)
```

2. Create and label 4 separate charts for the following equations (choose a range for x that makes sense):
    - y = sqrt(x)
    - y = x**3
    - y = 2**x
    - y = 1/(x+1)
    
### You can use the functions from the math module to help implement some of the equations above.

```python
import math
```

```python
n_rows = 2
n_cols = 2

x = pd.Series(range(0, 10))

y1 = x.apply(lambda x: math.isqrt(x))
y2 = x**3
y3 = 2**x
y4 = 1/(x+1)

plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y1)
plt.title('x ~ y1')

# the second subplot
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y2)
plt.title('x ~ y2')

plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y3)
plt.title('x ~ y3')

# the second subplot
plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y4)
plt.title('x ~ y4')

plt.tight_layout()
plt.suptitle('Different Ys')
plt.show()
```

4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.

```python
plt.figure(figsize=(12, 5))

n_rows = 2
n_cols = 2

x = pd.Series(range(0, 10))

y1 = x.apply(lambda x: math.isqrt(x))
y2 = x**3
y3 = 2**x
y4 = 1/(x+1)

plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y1, color='red')
plt.title('x ~ y1')

plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y2, color='green')
plt.title('x ~ y2')

plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y3, color='yellow')
plt.title('x ~ y3')

plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y4, color='purple')
plt.title('x ~ y4')


plt.tight_layout()
plt.suptitle('Different Ys')
plt.show()
```

```python
# Example given by Ryan

x = list(range(0, 25))

# generate y for each fucntion
y1 = [math.sqrt(n) for n in x]
y2 = [n**3 for n in x]
y3 = [2**n for n in x]
y4 = [1/(n+1) for n in x]

plt.plot(x, y1, c='#30a2da', label='$\sqrt{x}$', alpha=1.0)
plt.plot(x, y2, c='orange', label='$x^{3}$', alpha=.8)
plt.plot(x, y3, c= (1,0,1), label='$2^{x}$', alpha=.8)
plt.plot(x, y4, c= '#444444', label='$1/(x+1)$', alpha=.8)


plt.ylim(0, 10)
plt.xlim(0, 10)

plt.ylabel('$y$')
plt.xlabel('$x$')
plt.title('My Functions')
plt.legend()

plt.show()
```
